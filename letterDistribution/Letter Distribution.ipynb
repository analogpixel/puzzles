```python
#z =pd.DataFrame(index=['a','b','c'],columns=[1,2,3])

def getWordBreaks(wordlist):
    letters = [chr(x) for x in range(97,123)]
    z =pd.DataFrame(0, index=letters,columns=range(3,12))
    for words in wordlist.split(" "):
        l = len(words)
        if l < 3 or l > 11:
            continue
        for ch in words:
            if not ch in letters:
                continue               
            z.loc[ch][l] = z.loc[ch][l] + 1
    return z

wordData    = getWordBreaks("".join(open("applia.txt", encoding="utf8").readlines())  )
wordDataPuz = getWordBreaks("".join(open("puz.txt").readlines())  )
```

```python
wordData.sort_values(5).plot.bar(stacked=True, figsize=(20,10))
```

```python
def getBreak(wordlist):
    d = {}
    for letters in wordlist:
        letter = letters.lower()
        if letter in d:
                d[letter]+=1
        else:
                d[letter] = 1
    return d

ldist_p = getBreak(  "".join(open("alice.txt").readlines())  )
puzdist_p = getBreak(  "".join(open("puz.txt").readlines())  )
```

```python
ldist     = pd.Series(ldist_p).loc['a':'z'].sort_values()
puzdist   = pd.Series(puzdist_p).loc['a':'z'].sort_values()
```

```python
hist, bins = np.histogram(ldist, bins=50)
center = (bins[:-1] + bins[1:]) / 2
```
